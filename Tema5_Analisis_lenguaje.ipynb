# Preprocesamiento de texto

El preprocesamiento de texto en NLP es la etapa en la que un texto crudo se transforma en una versión más limpia y estructurada para que pueda ser analizada por un modelo. Este proceso suele incluir la normalización (como pasar todo a minúsculas, eliminar espacios extra o caracteres especiales), la limpieza de ruido (URLs, menciones, correos, hashtags, emojis), la tokenización en palabras u oraciones, la eliminación de stopwords, y la reducción de palabras a su forma base mediante lematización o stemming. También puede implicar el uso de expresiones regulares (regex) para detectar y reemplazar patrones como números, correos electrónicos o símbolos repetidos. Todo esto permite que el texto sea más uniforme y represente mejor la información antes de ser convertido en vectores o embeddings para el análisis.

NLTK (Natural Language Toolkit) es una librería de Python diseñada para trabajar con lenguaje natural, muy utilizada en la enseñanza y en proyectos académicos. Sirve para realizar tareas como tokenización, lematización, stemming, etiquetado gramatical, análisis de concordancias y manejo de corpus de texto. Además, incluye recursos lingüísticos listos para usar, como diccionarios, stopwords y datasets, lo que la hace ideal para experimentar y aprender los fundamentos del procesamiento de texto antes de pasar a librerías más rápidas o pensadas para producción como spaCy.

### Tokenizar

Tokenizar es el proceso de dividir un texto en unidades más pequeñas llamadas tokens.
Un token puede ser una palabra, un signo de puntuación, un número o incluso una oración, dependiendo de cómo se defina la segmentación.

```python
import nltk
from nltk.tokenize import word_tokenize, sent_tokenize

# Descargar recursos necesarios
# nltk.download("punkt")

texto = "El NLP es increíble. Me gusta aprender Python."

# Tokenización por palabras

tokens_palabras = word_tokenize(texto)
print("Tokens por palabras:", tokens_palabras)

# Tokenización por oraciones
tokens_oraciones = sent_tokenize(texto)
print("Tokens por oraciones:", tokens_oraciones)
```

```python
from bs4 import BeautifulSoup
 
import urllib.request
```

```python
url = "https://es.wikipedia.org/wiki/Finanzas"

# Agregar headers para simular un navegador
headers = {"User-Agent": "Mozilla/5.0"}

req = urllib.request.Request(url, headers=headers)

response = urllib.request.urlopen(req)  
# Abre la URL con los headers y obtiene la respuesta del servidor (Wikipedia sobre Finanzas)

html = response.read()  
# Lee el contenido HTML completo de la página web descargada

soup = BeautifulSoup(html, "html.parser")  
# Convierte el HTML en un objeto BeautifulSoup, lo que facilita analizarlo y manipularlo
# (ejemplo: quitar etiquetas, extraer párrafos, títulos, etc.)

text = soup.get_text(strip=True)  
# Extrae el texto limpio de la página, eliminando todas las etiquetas HTML
# El parámetro "strip=True" elimina espacios y saltos de línea innecesarios

print(text)  
# Muestra en consola solo los primeros 1000 caracteres del texto plano
```

```python
# Crear nuestros tokens a partir de un texto

# Tokenizamos el texto: divide el texto en palabras y signos de puntuación
tokens = word_tokenize(text)  

# Filtramos los tokens:
# 1. Convertimos cada palabra a minúsculas (para uniformidad).
# 2. Usamos isalpha() para quedarnos solo con tokens que contengan letras (así eliminamos números y signos de puntuación).
tokens = [word.lower() for word in tokens if word.isalpha()]  

# Imprimimos la lista final de tokens limpios
print(tokens)  
```

```python
nltk.FreqDist(tokens)
```

```python
# Crear la distribución de frecuencias
freq = nltk.FreqDist(tokens)

# Mostrar palabras ordenadas de mayor a menor frecuencia
for key, val in freq.most_common():
    print(str(key) + ": " + str(val))
```

### Palabras de parada

Las palabras de parada o stopwords son aquellas palabras muy comunes en un idioma que, por sí solas, aportan poco significado semántico al análisis de texto. Ejemplos en español son: “el”, “la”, “de”, “y”, “que”, “en”, etc.

### Por qué son importantes?

* Aparecen con mucha frecuencia, lo que puede enmascarar patrones útiles si no se manejan adecuadamente.

* Su presencia puede inflar estadísticas de frecuencia y desviar el análisis hacia palabras que realmente no aportan información relevante.  

### Cuándo conviene quitarlas  

* En tareas como análisis de frecuencia, clasificación de texto, vectorización (BoW o TF-IDF) o análisis de sentimiento, donde queremos resaltar palabras que aportan significado al texto, suele ser conveniente eliminarlas.

### No conviene quitarlas cuando:
* Se hace análisis gramatical 

* Se necesita conservar contexto completo, por ejemplo en tareas de generación de texto o modelos basados en transformers, ya que algunas stopwords ayudan a mantener la estructura de las oraciones.

En Python, NLTK y spaCy tienen listas predefinidas de stopwords, y también se pueden crear listas personalizadas según el dominio del texto (por ejemplo, eliminar palabras muy frecuentes en correos corporativos).

### Derivación Regresiva

En muchos contextos de NLP o lingüística computacional se refiere a la técnica de reducir una palabra a su raíz o forma base eliminando sufijos o terminaciones. Esto se conoce más comúnmente como stemming.  

Por ejemplo, en español:  

“corriendo”, “corrí”, “correrán” → raíz: “corr”  

“jugaré”, “jugando”, “jugador” → raíz: “jug”

#### Para qué se usa

* Se utiliza principalmente en procesamiento de texto y recuperación de información:

* Búsqueda de información: permite que al buscar “correr”, también se encuentren documentos que contienen “corriendo” o “corrí”.

* Clasificación de texto y análisis de frecuencia: agrupa palabras con la misma raíz para evitar contar variantes como palabras distintas.

* Reducción de dimensionalidad en NLP: disminuye el tamaño del vocabulario en modelos de Bag-of-Words o TF-IDF.

#### Cuándo es útil

* Cuando se busca simplificar el texto sin perder el sentido básico de las palabras.

* Para textos con muchas variantes de una misma palabra, como blogs, noticias o redes sociales.

* No es tan útil cuando se necesita precisión gramatical, como análisis sintáctico o generación de texto, porque el stemming puede eliminar sufijos importantes y producir raíces no reconocibles (“corr” en lugar de “correr”).

```python
from nltk.stem.snowball import SnowballStemmer
from nltk.stem import PorterStemmer 
```

###### PorterStemmer

* Fue diseñado originalmente para inglés.

* Es simple y rápido, pero a veces demasiado agresivo en la reducción de palabras.

```python
# Crear un objeto Stemmer usando el algoritmo de Porter
stemmer = PorterStemmer()

# Aplicar stemming a diferentes formas de la palabra "play"
print(stemmer.stem('play'))     # Salida: 'play'
print(stemmer.stem('played'))   # Salida: 'play'
print(stemmer.stem('playing'))  # Salida: 'play'
print(stemmer.stem('player'))  # Salida: 'play'
```

##### SnowballStemmer

* Es más moderno y soporta varios idiomas (incluido español).

* Menos agresivo y más preciso que Porter en muchos casos.

```python
from nltk.stem.snowball import SnowballStemmer

# Crear un stemmer para español
spanish_stemmer = SnowballStemmer('spanish')

# Palabras para aplicar stemming
words = ["trabajando", "trabajo", "trabajaría", "trabajadores"]

# Aplicar stemming
for word in words:
    print(f"{word} -> {spanish_stemmer.stem(word)}")
```

```python
# Crear un stemmer para inglés
english_stemmer = SnowballStemmer('english')

# Palabras para aplicar stemming
words = ["playing", "played", "plays", "player"]

# Aplicar stemming
for word in words:
    print(f"{word} -> {english_stemmer.stem(word)}")
```

### Palabras lematizadoras

La lematización es el proceso de reducir una palabra a su forma base o lema, que es una palabra real y reconocible en el idioma.  
  
A diferencia del stemming, que solo corta sufijos de manera aproximada, la lematización produce palabras completas correctas.

Ejemplo en español:  

“corriendo” → “correr”  

“trabajaría” → “trabajar”  

“mejores” → “mejor”  

En inglés:  

“running” → “run”  

“better” → “good”  

##### Para qué sirve  

La lematización se usa para:  

* Normalizar el texto: todas las formas de una palabra se convierten en su lema, evitando duplicados en el análisis.

* Mejorar modelos de NLP: reduce el tamaño del vocabulario en modelos Bag-of-Words, TF-IDF o embeddings.

* Análisis semántico más preciso: mantiene la forma real de la palabra, lo que permite interpretar correctamente los resultados.

* Búsqueda y recuperación de información: encontrar documentos que contengan diferentes formas de una misma palabra.

##### Cuándo se usa

* Cuando se necesita precisión semántica y queremos que la raíz sea una palabra válida del idioma.

* Clasificación de texto

* Análisis de sentimientos

* Resúmenes automáticos

* Chatbots y asistentes virtuales

* No es estrictamente necesario en tareas muy rápidas de conteo de frecuencia o clustering, donde stemming puede ser suficiente y más rápido.

##### Resumen práctico:

* Stemming → rápido, raíces aproximadas, puede producir palabras no reales.

* Lematización → más preciso, palabras reales, más costoso computacionalmente.

```python
from nltk.stem import WordNetLemmatizer
```

```python
# Crear un objeto lematizador de NLTK
lemmatizer = WordNetLemmatizer()

# Lematizar la palabra 'working' como verbo
# Por defecto, lemmatize() asume que la palabra es un sustantivo
# Al pasar pos="v" indicamos que es un verbo, así devuelve su forma base correcta

print(lemmatizer.lemmatize('working', pos="v"))  # Salida: 'work'
```

# Word embedings

Un word embedding es una representación numérica de las palabras en un espacio vectorial. Cada palabra se convierte en un vector de números, de manera que palabras con significados similares quedan cercanas entre sí en ese espacio. Por ejemplo, "king" y "queen" estarán más cerca que "king" y "banana". Esto permite que los algoritmos puedan “entender” relaciones semánticas y sintácticas entre palabras.

Los embeddings se utilizan en casi todas las tareas modernas de procesamiento de lenguaje natural. Sirven para clasificación de texto (por ejemplo, detectar spam o analizar sentimientos), búsqueda semántica de documentos relacionados, traducción automática, construcción de chatbots, y también para visualización o clustering de palabras según su significado. Básicamente, permiten que los modelos trabajen con significado y contexto de palabras, no solo con su forma.

Se usan siempre que queremos que un modelo de NLP comprenda el significado y contexto de las palabras en lugar de tratarlas como simples símbolos. Son especialmente útiles cuando se trabaja con grandes corpus de texto o cuando la relación semántica entre palabras es importante para la tarea, como en modelos de clasificación, generación de texto, traducción automática o sistemas de recomendación de contenido textual.

#### Como se hacen?  

Los word embeddings pueden obtenerse de dos formas: usando modelos pre-entrenados o entrenándolos en tu propio corpus. Los pre-entrenados, como GloVe, Word2Vec, fastText o los modelos medianos de spaCy (en_core_web_md, es_core_news_md), ya vienen con vectores listos para muchas palabras comunes. Si tu corpus es muy específico, puedes entrenar tus propios embeddings usando librerías como gensim, de manera que los vectores reflejen mejor el significado de tu vocabulario particular.

## Embeddings pre entrenados

Un embedding preentrenado es un conjunto de vectores de palabras ya entrenados en un corpus grande y diverso, como Wikipedia, noticias o Common Crawl.  

Cada palabra tiene asociado un vector numérico que captura su significado y relaciones con otras palabras.  

Lo “preentrenado” significa que otra persona o equipo ya entrenó el modelo usando técnicas como Word2Vec, GloVe o FastText, por lo que no necesitas entrenarlo tú mismo.  

#### Por qué se usa?

* Ahorras tiempo y recursos: no necesitas millones de frases ni potencia de cómputo.

* Obtienes vectores de alta calidad para palabras comunes y muchas relaciones semánticas.

* Ideal cuando tu corpus es pequeño o no tiene suficiente contexto para entrenar un Word2Vec desde cero.

La diferencia principal entre Skip-gram y CBOW radica en la dirección de predicción durante el entrenamiento de Word2Vec. Skip-gram toma una palabra central (target) y trata de predecir las palabras que aparecen a su alrededor (contexto), mientras que CBOW hace lo contrario: toma las palabras del contexto y predice la palabra central. Skip-gram es más útil cuando se trabaja con textos pequeños o con palabras poco frecuentes, ya que puede capturar relaciones semánticas finas incluso con pocas apariciones. Por otro lado, CBOW suele ser más rápido y eficiente en textos grandes, aprendiendo de manera más estable la representación de palabras frecuentes. En general, se usa Skip-gram cuando se quiere priorizar la calidad de embeddings para vocabularios raros o específicos, y CBOW cuando se necesita velocidad y se trabaja con textos extensos y vocabularios más comunes.

Ahora, construiremos una conversación sencilla con un “tutor de IA” para mostrar cómo el modelo mantiene el contexto entre turnos. Se inicia con una pregunta y posteriormente se realiza una consulta de seguimiento, evidenciando que el modelo recuerda la interacción previa.

La nueva API de OpenAI simplifica el manejo del contexto conversacional al evitar el envío completo del historial en cada solicitud. En su lugar, se utiliza el parámetro previous_response_id, que enlaza cada nueva petición con la respuesta anterior, permitiendo que la API gestione el contexto automáticamente.

La conversación se estructura a partir de tres elementos principales: instructions, que define el comportamiento general del modelo a lo largo de la interacción; input, que corresponde a la entrada del usuario en cada turno; y previous_response_id, que conecta los turnos sucesivos. Este enfoque es más eficiente, ya que reduce la cantidad de datos enviados y, por lo tanto, el consumo de tokens en conversaciones largas.

```python
print("\n--- Starting Conversation: Turn 1 ---")

# Define the system message (persona) for the AI Tutor
system_instructions = "You are a helpful AI Tutor explaining Large Language Model concepts simply."

# Define the user's first question
user_input_1 = "Can you explain what 'tokens' are in the context of LLMs, like I'm new to this?"

print(f"System Instructions: {system_instructions}")
print(f"User Input: {user_input_1}")

# Define parameters for this call
MODEL = "gpt-5-mini"
MAX_OUTPUT_TOKENS=500 # (GPT 5 & GPT 5 mini are reasoning model, so includes visible output tokens and reasoning tokens)
REASONING_EFFORT ="minimal" # Constrains effort on reasoning for reasoning models. Currently supported values are minimal, low, medium, and high
```

```python
try:
    print(f"\nMaking API call to {MODEL}...")
    # Use the Response API
    response_1 = client.responses.create(
        model=MODEL,
        instructions=system_instructions,
        input=user_input_1,
        max_output_tokens=MAX_OUTPUT_TOKENS,
        reasoning= {'effort':REASONING_EFFORT},
    )
    print("API call successful.")
    # --- Process the response from the first turn ---
    # Extract the assistant's reply content
    assistant_response_1 = response_1.output[1].content[0].text

    # Save the response ID for conversation continuity
    response_id_1 = response_1.id

    print("\nAI Tutor (Turn 1):")
    print(assistant_response_1)

    # Print token usage for this call
    usage_1 = response_1.usage
    print("\n\n===========================================")
    print(f"\nToken Usage (Turn 1): Input={usage_1.input_tokens}, Output={usage_1.output_tokens}, Total={usage_1.total_tokens}")

except openai.APIError as e:
    print(f"OpenAI API returned an API Error: {e}")
except openai.AuthenticationError as e:
    print(f"OpenAI Authentication Error: {e}")
except Exception as e:
    print(f"An unexpected error occurred: {e}")
```

Para realizar una pregunta de seguimiento, únicamente se envía la nueva entrada del usuario junto con el identificador (response_id) obtenido en la interacción anterior. De esta forma, no es necesario construir ni reenviar manualmente el historial completo de la conversación. El uso del parámetro previous_response_id permite enlazar los turnos, indicando a la API que la nueva consulta es una continuación directa de la respuesta previa.

```python
print("\n--- Continuing Conversation: Turn 2 ---")

# Define the user's second question
user_input_2 = "Thanks! So, based on your explanation, are common words like 'the' or 'is' usually single tokens?"

print(f"\nUser Input (Turn 2): {user_input_2}")
```

```python
try:
    print(f"\nMaking API call to {MODEL} (Turn 2)...")
    response_2 = client.responses.create(
        model=MODEL,
        instructions=system_instructions,
        input=user_input_2,
        max_output_tokens=MAX_OUTPUT_TOKENS,
        reasoning= {'effort':REASONING_EFFORT},
        previous_response_id=response_id_1  # Link to previous response for context
    )
    print("API call successful.")

    # --- Process the response from the second turn ---
    assistant_response_2 = response_2.output[1].content[0].text

    print("\nAI Tutor (Turn 2):")
    print(assistant_response_2)

    # Print token usage for this call
    usage_2 = response_2.usage

    print("\n\n===========================================")
    print(f"\nToken Usage (Turn 2): Input={usage_2.input_tokens}, Output={usage_2.output_tokens}, Total={usage_2.total_tokens}")

except openai.APIError as e:
    print(f"OpenAI API returned an API Error: {e}")
except openai.AuthenticationError as e:
    print(f"OpenAI Authentication Error: {e}")
except Exception as e:
    print(f"An unexpected error occurred: {e}")
```

El parámetro reasoning permite controlar la profundidad del proceso de razonamiento interno del modelo antes de generar una respuesta. Mediante el ajuste del nivel de effort, es posible indicar cuántos recursos computacionales debe emplear el modelo, desde tareas simples hasta problemas que requieren razonamiento complejo y en múltiples pasos.

Los niveles de esfuerzo van desde minimal, adecuado para tareas rápidas y sencillas, hasta high, reservado para problemas que demandan análisis profundo, deducción lógica o creatividad avanzada. Sin embargo, es importante considerar que un mayor esfuerzo implica un mayor consumo de tokens y un incremento en el tiempo de respuesta y el costo de la API, aun cuando esos pasos internos no siempre sean visibles en la salida final.

Por ello, se recomienda comenzar con niveles bajos de esfuerzo y aumentarlos únicamente si la calidad de la respuesta no es suficiente, logrando así un equilibrio entre precisión y costo computacional.

El prompting es la habilidad fundamental para comunicarse de manera efectiva con los Modelos de Lenguaje de Gran Escala (LLMs), ya sea al utilizarlos directamente o al construir productos sobre ellos. Consiste en formular instrucciones claras y precisas que permitan al modelo comprender qué tarea debe realizar y cómo debe responder, actuando como la interfaz entre la intención humana y la salida generada por la IA.

Un prompt es la entrada que se proporciona al modelo, la cual puede presentarse como una pregunta, una descripción de tarea o un conjunto de ejemplos e instrucciones. La calidad de la respuesta depende directamente de la calidad del prompt; instrucciones ambiguas o mal diseñadas suelen producir resultados deficientes, siguiendo el principio de “garbage in, garbage out”.

El prompt engineering ha adquirido relevancia como una habilidad profesional, aunque no requiere conocimientos avanzados en aprendizaje automático, sino práctica, sentido común y comprensión del contexto del problema. En aplicaciones reales, suele ser necesario integrar conocimiento del dominio para diseñar prompts adaptados a flujos de trabajo específicos y a las necesidades del usuario final, además de reducir posibles fallos del modelo.

Finalmente, el prompting debe entenderse como un proceso iterativo: las técnicas pueden cambiar y no siempre funcionan perfectamente en todos los casos, por lo que es necesario probar, evaluar y refinar los prompts de forma continua. En esencia, esta disciplina se basa en desarrollar intuición para interactuar con los LLMs mediante lenguaje natural y obtener resultados más útiles y consistentes.

# Técnicas de prompt

Las técnicas de prompting son estrategias para formular instrucciones efectivas a los modelos de lenguaje, con el fin de obtener respuestas más precisas, útiles y alineadas con un objetivo específico. Una de las formas más simples es el zero-shot prompting, que consiste en realizar una petición directa sin proporcionar ejemplos ni contexto adicional. Este enfoque puede funcionar bien para tareas sencillas, aunque su calidad es variable.

Cuando se requiere mayor control, se utiliza el few-shot prompting o in-context learning, donde se incluyen ejemplos que guían al modelo sobre el formato, el estilo o el tipo de razonamiento esperado. Esta técnica permite enseñar al modelo “por demostración” y mejora significativamente la consistencia de las respuestas, especialmente en tareas de resumen, clasificación o transformación de texto.

Otra estrategia es el role prompting, que asigna al modelo un rol o perspectiva específica (por ejemplo, profesor, analista legal o divulgador). Al definir una identidad, se condiciona implícitamente el tono, el nivel de detalle y el enfoque de la respuesta, manteniendo el contenido factual pero adaptando su presentación.

El chain-of-thought prompting busca mejorar el desempeño en problemas complejos solicitando explícitamente que el modelo razone paso a paso antes de dar una respuesta final. Este enfoque es especialmente útil en tareas que requieren razonamiento lógico, matemático o simbólico, ya que reduce errores al obligar al modelo a descomponer el problema en etapas intermedias. No obstante, su efectividad depende del tipo de tarea y del modelo utilizado.

Finalmente, el prompting es un proceso iterativo. Rara vez se obtiene la mejor respuesta en el primer intento; por ello, es común ajustar y refinar los prompts a partir de las respuestas generadas. Cada salida del modelo sirve como retroalimentación para mejorar la instrucción, permitiendo afinar progresivamente el resultado hasta que cumpla con los requisitos deseados.

## Tips para la creación de prompts

* Formular instrucciones claras, precisas y explícitas es clave para obtener respuestas más exactas del modelo.

* El prompt engineering es un proceso iterativo, que implica probar, evaluar y refinar los prompts de manera continua.

* Es importante considerar escenarios poco comunes y no solo los casos típicos al diseñar un prompt.

* Los prompts deben probarse con entradas imperfectas o desordenadas, similares a las que podrían proporcionar usuarios reales.

* Analizar cuidadosamente las respuestas del modelo permite verificar si sigue las instrucciones y detectar posibles ambigüedades.

* No se debe asumir información implícita; es recomendable descomponer las tareas y proporcionar todo el contexto necesario.

* Es fundamental reflexionar sobre cómo el modelo interpreta las instrucciones, ya que puede entenderlas de forma distinta a la intención original.

* La creación de prompts debe tratarse como un proceso de ingeniería, utilizando prácticas como control de versiones y experimentación sistemática.

* Solicitar al propio modelo que identifique partes ambiguas o poco claras del prompt puede ayudar a mejorarlo.

* Mantener los prompts claros y concisos, evitando complejidad innecesaria.

* Equilibrar la atención entre casos normales y casos límite, sin perder de vista el objetivo principal.

* Considerar el sistema completo en el que se integra el prompt, incluyendo datos, latencia e interacción con otros componentes.

* Reconocer que el prompt engineering combina comunicación efectiva y pensamiento sistemático, no solo habilidades de redacción.

* Estudiar de forma continua los outputs del modelo para mejorar la intuición y la calidad de futuras interacciones.

#### Prompt Injection y Prompt Hacking: visión general

Los Modelos de Lenguaje de Gran Escala (LLMs) enfrentan riesgos importantes de seguridad, costos y confiabilidad, entre los cuales destaca el prompt injection. Este tipo de amenaza explota la forma en que los modelos interpretan instrucciones en lenguaje natural, permitiendo que usuarios maliciosos influyan en el comportamiento del modelo más allá de lo previsto por sus diseñadores. Estas vulnerabilidades pueden afectar tanto a aplicaciones individuales como a la confianza general en los sistemas de IA, especialmente en dominios sensibles como educación, salud o finanzas.

#### ¿Qué es Prompt Injection?

El prompt injection es una técnica de manipulación mediante la cual se diseñan entradas específicas para provocar respuestas no deseadas o comportamientos indebidos en un LLM. Estas manipulaciones pueden ir desde intentos benignos de “engañar” al modelo hasta ataques más graves que buscan acceder a información sensible, eludir restricciones de seguridad o utilizar el sistema fuera de su propósito original.

El riesgo del prompt injection no se limita a respuestas incorrectas; también puede comprometer la integridad de los datos, la privacidad de los usuarios y el cumplimiento de normas éticas y regulatorias. Por ello, comprender este fenómeno es fundamental para garantizar un uso seguro y responsable de los LLMs en aplicaciones reales.

#### Tipos de Prompt Injection  

Existen dos categorías principales de ataques de prompt injection: directos e indirectos.  

#### Prompt Injection Directo

El prompt injection directo ocurre cuando un atacante introduce instrucciones explícitas en la entrada del modelo con el objetivo de manipular su respuesta. Estas instrucciones pueden estar ocultas, disfrazadas o formuladas de manera engañosa para inducir al modelo a ignorar sus reglas internas.

Un ejemplo común es el llamado “jailbreaking”, donde se utilizan prompts diseñados para desactivar mecanismos de seguridad y forzar al modelo a generar contenido restringido o revelar información confidencial. Este tipo de ataque es relativamente fácil de ejecutar y representa una amenaza directa para sistemas que no cuentan con validaciones adecuadas de entrada.

#### Prompt Injection Indirecto

El prompt injection indirecto es más sutil y potencialmente más peligroso. En este caso, el atacante no modifica directamente la entrada del usuario, sino que altera el contexto externo del que el modelo obtiene información. Esto ocurre, por ejemplo, cuando un LLM consume datos de páginas web, documentos o bases de conocimiento externas.

Un atacante puede insertar instrucciones ocultas dentro de estos recursos externos, de modo que cuando el modelo los procese (por ejemplo, al resumir una página web), termine siguiendo directrices maliciosas. Este tipo de ataque es especialmente relevante en sistemas que integran LLMs con herramientas de recuperación de información o navegación web.

#### ¿Qué es Prompt Leakage?

El prompt leakage se produce cuando un LLM revela total o parcialmente sus prompts internos, incluidos system prompts o instrucciones confidenciales. Esta filtración puede exponer lógica propietaria, reglas de seguridad internas o información sensible utilizada para guiar el comportamiento del modelo.

Un caso documentado ocurrió cuando usuarios lograron extraer prompts internos de un motor de búsqueda basado en ChatGPT mediante consultas cuidadosamente diseñadas. Este tipo de vulnerabilidad representa un riesgo adicional, ya que facilita ataques posteriores y compromete la propiedad intelectual y la seguridad del sistema.

#### Mitigación de riesgos de Prompt Injection y Prompt Leakage

Para reducir los riesgos asociados al prompt injection y al prompt leakage, pueden implementarse diversas estrategias complementarias:  

* Validación robusta de entradas, mediante filtros que detecten instrucciones sospechosas o patrones maliciosos.

* Control estricto de fuentes externas, asegurando que los modelos solo interactúen con contenidos verificados y confiables.

* Aislamiento entre instrucciones del sistema y entradas del usuario, evitando que estas últimas puedan sobrescribir reglas internas.

* Educación y concientización de los usuarios, promoviendo un uso responsable y seguro de los modelos de lenguaje.

* Monitoreo y evaluación continua, para detectar comportamientos anómalos y ajustar los mecanismos de defensa.

* Estas medidas son esenciales para mantener la seguridad, confiabilidad y uso ético de los LLMs, especialmente en sistemas desplegados a gran escala o en contextos críticos.

## Evaluación de prompts

Una suposición común al trabajar con modelos de lenguaje es que un prompt bien diseñado producirá resultados similares en distintos LLMs; sin embargo, en la práctica esto no siempre ocurre. Aunque los modelos puedan haber sido entrenados con datos similares, diferencias en su arquitectura, métodos de entrenamiento y procesos de ajuste fino hacen que cada uno interprete y responda a los prompts de manera distinta. Por ello, un prompt optimizado para un modelo puede no ser igual de efectivo en otro.

La gestión de la ventana de contexto (prompt window management) también influye en el comportamiento del modelo, ya que se refiere a cómo se estructura y optimiza la información de entrada que define el contexto de la respuesta. Una mala organización del prompt puede dificultar que el modelo comprenda correctamente la tarea, aun cuando la instrucción sea conceptualmente adecuada.

En el caso de los LLMs propietarios (como OpenAI, Gemini o Claude), las actualizaciones frecuentes pueden mejorar capacidades como el razonamiento o la alineación, pero también alterar el comportamiento de prompts previamente optimizados. Por ello, las aplicaciones que dependen de estos modelos deben asumir que su comportamiento puede cambiar con el tiempo y requieren monitoreo continuo.

Por otro lado, los modelos de código abierto ofrecen mayor estabilidad, ya que su comportamiento permanece constante mientras se use la misma versión del modelo. Esto resulta ventajoso en aplicaciones donde la consistencia es prioritaria, aunque implica que las mejoras, correcciones y mantenimiento recaen en el usuario.

Dado este contexto, es fundamental evaluar periódicamente el desempeño de los prompts, utilizando métricas relevantes para cada aplicación. Una estrategia efectiva consiste en emplear otro modelo como evaluador (LLM-as-a-judge) para verificar si las respuestas cumplen criterios específicos de formato, estilo y contenido. La evaluación continua y la optimización iterativa de prompts permiten mantener la calidad y robustez de los sistemas basados en LLMs, independientemente del tipo de modelo utilizado.

```python
prompt_generation = """
Generate a table with the 5 most popular pharmaceutical companies and their foundation years.
The response should include only the table, with no additional text.
Use the following example format:
---
Company | Foundation Year
Microsoft | 1975
Apple | 1976
Google | 1998
Amazon | 1994
Meta | 2004
---"""


# Making the API call
response = client.responses.create(
  model="gpt-5",
  instructions="You are a helpful assistant.",
  input=prompt_generation,
  reasoning={'effort':'minimal'},
)

generated_response = response.output_text
print(generated_response)
```

La técnica LLM as a Judge consiste en utilizar un modelo de lenguaje, como GPT-5, para evaluar si una salida cumple con criterios o formatos específicos. En este caso, el modelo actúa como evaluador verificando si una tabla se ajusta a una estructura previamente definida en el prompt. La entrada proporcionada al modelo incluye únicamente las instrucciones del formato esperado y la tabla a evaluar, sin información adicional, con el fin de garantizar una validación objetiva. Este enfoque permite automatizar la revisión de resultados y sirve como base para evaluaciones más avanzadas dentro de arquitecturas como RAG.

```python
prompt_check_table_new = """Your task is to verify if a given table matches the exact format and structure of a provided example table.

Here's an example of the format that the table should have:
---
Company | Foundation Year
Microsoft | 1975
Apple | 1976
Google | 1998
Amazon | 1994
Meta | 2004
---

Table to Check:
{table_to_check}

Instructions:
- The table to check should match the structure, headers, and format of the Example Table exactly.
- The column names must be "Company" and "Foundation Year".
- The values in each row should have the company names and their corresponding foundation years.
- If the given table matches the example table in all these aspects, write "Yes".
- Write "No" if there are any differences in structure, headers, or if any company/year is missing or incorrect.

Only respond with "Yes" or "No".

"""

formatted_input = prompt_check_table_new.format(table_to_check=generated_response)
print(formatted_input)
```

```python
# using LLM as a Judge to check the format
response = client.responses.create(
  model="gpt-5",
  instructions="You are a strict judge. Evaluate inputs based on the given criteria and provide only the required response",
  input=formatted_input,
  reasoning={'effort':'minimal'},
)
print(response.output_text)
```

### Introduccion a RaG

La Generación Aumentada por Recuperación (RAG), al integrarse con modelos de lenguaje de gran escala como GPT-5, permite mejorar significativamente la precisión y confiabilidad de las respuestas generadas. Al incorporar información externa verificada —por ejemplo, documentos o publicaciones—, RAG reduce la probabilidad de alucinaciones y produce respuestas más fundamentadas y alineadas con datos reales, lo que resulta especialmente valioso en aplicaciones prácticas.

El funcionamiento de RAG se basa en recuperar información relevante de un corpus o base de datos durante el proceso de generación, utilizando dicha información para enriquecer el prompt del modelo. Un reto central de este enfoque consiste en seleccionar y filtrar cuidadosamente los fragmentos de texto más relevantes, de modo que se maximice su utilidad dentro de la ventana de contexto limitada del modelo. Esta optimización permite reducir costos y latencia al minimizar el número de tokens procesados y evita introducir información irrelevante o confusa.

En términos generales, un sistema RAG se compone de tres elementos principales: un mecanismo de recuperación de documentos relevantes, una etapa de augmentación que combina dichos documentos con la consulta del usuario, y un modelo de lenguaje encargado de generar la respuesta final. Este enfoque aprovecha la capacidad de los LLMs para aprender a partir del contexto y adaptarse a distintos formatos, tareas y est

# Arquitectura 

En una arquitectura de Retrieval-Augmented Generation (RAG), el componente de recuperación desempeña un papel fundamental, ya que su función principal es interpretar la consulta del usuario y localizar información semánticamente relevante. Para ello, se emplean modelos de embeddings, que codifican tanto las fuentes de conocimiento (como bases de datos o documentos PDF) como la consulta del usuario en representaciones vectoriales. Posteriormente, se aplica una métrica de similitud —comúnmente la similitud del coseno— para identificar los documentos o fragmentos que mejor se alinean con la intención de la consulta. Este componente también decide cuántos documentos recuperar y, en algunos casos, qué secciones específicas resultan más pertinentes.

La selección del modelo de lenguaje (LLM) es igualmente crítica, ya que este se encarga de procesar el prompt enriquecido y sintetizar la respuesta final. Dependiendo de la aplicación, puede ser necesario utilizar modelos especializados, como modelos orientados a programación para consultas de código. Además, es indispensable que el modelo cuente con una ventana de contexto lo suficientemente amplia para procesar toda la información relevante. En consultas complejas, esto puede implicar integrar múltiples documentos para permitir razonamientos que combinen información proveniente de distintas fuentes.

La ventana de contexto determina el número máximo de tokens que un modelo puede procesar simultáneamente, variando desde unos pocos miles en modelos antiguos hasta cientos de miles en versiones más recientes. Aumentar esta capacidad implica mayores costos computacionales y consumo de hardware, por lo que se requieren estrategias para optimizar el uso del contexto. Entre ellas se encuentran técnicas como la resumización de documentos, que permiten condensar la información sin perder contenido esencial.

Para mejorar la eficiencia y efectividad de los sistemas RAG en entornos productivos, se emplean diversas estrategias, como el chunking (división de documentos extensos en fragmentos manejables), la anotación con metadatos relevantes (por ejemplo, palabras clave o categorías), y el diseño cuidadoso de la augmentación del prompt. Estas técnicas facilitan una recuperación más precisa y reducen el ruido informativo.

Finalmente, el diseño de sistemas RAG y pipelines basados en LLMs requiere integrar conocimiento experto del dominio. Más allá de los aspectos técnicos, es crucial comprender los datos, los problemas específicos del sector y el contexto en el que se utilizan las respuestas generadas. La falta de esta comprensión puede conducir a errores sutiles pero críticos, como recuperar información de periodos o divisiones incorrectas en aplicaciones financieras. Por ello, combinar habilidades técnicas con conocimiento del dominio y criterios humanos se vuelve un factor clave para construir sistemas RAG confiables y eficaces.

* client.models.generate_content(...)

Ejecuta una solicitud al modelo Gemini para generar contenido.

Es el punto central donde se define cómo y qué responde el modelo.



* model

Especifica el modelo de Gemini que se va a utilizar.

Determina el equilibrio entre velocidad, costo y capacidad de razonamiento.



* contents

Contiene la entrada principal del usuario.

Incluye la consulta, el contexto y cualquier información adicional necesaria para generar la respuesta.



* config

Agrupa los parámetros que controlan el comportamiento del modelo durante la generación.



* thinking_config

Define el nivel de razonamiento interno permitido al modelo.

Limitarlo reduce costo y latencia; aumentarlo permite razonamiento más profundo.



* thinking_budget

Indica cuántos recursos de razonamiento puede usar el modelo.

Un valor bajo prioriza respuestas rápidas y deterministas.



* system_instruction

Establece reglas globales que guían el tono, formato y restricciones de la respuesta.

Se utiliza para alinear el comportamiento del modelo con el objetivo de la aplicación.



* temperature

Controla el grado de aleatoriedad en la generación.

Valores bajos producen respuestas más consistentes y reproducibles.
